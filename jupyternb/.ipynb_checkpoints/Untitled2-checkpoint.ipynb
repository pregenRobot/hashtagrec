```python
import csv

with open("mined_data/data_X.csv","rt") as f:
        reader = csv.reader(f)
        X = list(reader)
        
with open("mined_data/data_y.csv","rt") as f:
        reader = csv.reader(f)
        y = list(reader)
        
X = np.array(X)

y = np.array(y)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
import numpy as np
import copy
```

```python
import pandas as pd

splitter = StratifiedShuffleSplit(n_splits=1,random_state=13,test_size=0.2)

column_labels = list(range(50))
column_labels = [str(i) for i in column_labels]
column_labels.append("tag")

data = np.hstack((X,y))
dataDF = pd.DataFrame(data,columns=column_labels)

for train_index, test_index in splitter.split(dataDF,dataDF["tag"]):
    stratified_train = dataDF.loc[train_index]
    stratified_test = dataDF.loc[test_index]
    
X_train = stratified_train[column_labels[:-1]]
y_train = stratified_train["tag"]
```

```python
from sklearn.linear_model import SGDClassifier

sgdclassifier = SGDClassifier(random_state=13)

sgdclassifier.fit(X_train,y_train)
```

```python
X_test = stratified_test[column_labels[:-1]]
y_test = stratified_test["tag"]

sgdclassifier.score(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(sgdclassifier,X_train,y_train,cv=3,scoring="accuracy")
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

y_train_predictions = cross_val_predict(sgdclassifier,X_train,y_train,cv=3)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

cfmatrix = confusion_matrix(y_train,y_train_predictions)

plt.imshow(cfmatrix)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtreeclassifier = DecisionTreeClassifier()

y_train_predictions_dtree = cross_val_predict(dtreeclassifier,X_train,y_train,cv=3)

cfmatrix_dtree = confusion_matrix(y_train,y_train_predictions_dtree)

plt.imshow(cfmatrix_dtree)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knclassifier = KNeighborsClassifier()

y_train_predictions_kn = cross_val_predict(knclassifier,X_train,y_train,cv=3)
cfmatrix_kn = confusion_matrix(y_train,y_train_predictions_kn)

plt.imshow(cfmatrix_kn)
```

```python
##Precision -- Out of all POSITIVELY CLASSIFIED, HOW MANY ARE ACTUALLY POSITIVE? -- TP/(TP + FP) 
##Recall -- Out of all POSSIBLE POSITIVES, HOW MANY POSITIVES WERE CLASSIFIED CORRECTLY? -- TP/(TP+FN)
##Sensitivity
##True Positive Rate(TPR)
##This project should prioritize recall

# from sklearn.metrics import precision_score, recall_score

# precision_score(y_train,y_train_predictions_kn)

y_train_predictions_kn
```

```python
def calculate_scores(confusionmatrix):
    recalls = []
    precisions = []
    
    for i,row in enumerate(confusionmatrix):
        precisions.append(confusionmatrix[i][i]/np.sum(row))
        
    confusionmatrix =  np.array(confusionmatrix).T
    
    for i,column in enumerate(confusionmatrix):
        recalls.append(confusionmatrix[i][i]/np.sum(column))
        
    return recalls,precisions


recall, precision = calculate_scores(cfmatrix_kn)
      
print(np.average(recall))
print(np.average(precision))
```
